```python
import sqlite3
import numpy as np
import pandas as pd
import glob
import datetime
# modellen importeren
from sklearn.cluster import AgglomerativeClustering
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df =  pd.read_csv('beweging.csv')
```

```python
df = df.drop('Unnamed: 0', axis = 1)
```

```python
df.Tijd = pd.to_datetime(df.Tijd, dayfirst= True)
```

```python
df.head()
```

```python
actief = df[df.Omschrijving == 'Brug Beweging Actief']
sluiten = df[df.Omschrijving == 'Algemeen Commando Brug Sluiten']
openen = df[df.Omschrijving == 'Algemeen Commando Brug Openen'] 
```

```python
actief.info()
```

```python
begin = 0
eind = 0
tst = None
sluit = True
duur = list()
oslist = list()
timestamp = list()

for index, row in df.iterrows():
    waarde = row[2]
    oms = row[3]
    sec = row[4]
    ts = row[1]
    if ((oms == 'Brug Beweging Actief') & (waarde == True)):
        begin = sec
        tst = ts 
    elif ((oms == 'Brug Beweging Actief') & (waarde == False)):
        eind = sec
    if (oms == 'Algemeen Commando Brug Openen'):
        sluit = False
    elif (oms == 'Algemeen Commando Brug Sluiten'):
        sluit = True
        
    if ((eind!=0) & (begin!=0)):
        duur.append(eind-begin)
        timestamp.append(ts)
        oslist.append(sluit)
        eind = 0
        begin = 0

duurdf  = pd.DataFrame(duur)
duurdf['time'] = timestamp
duurdf['sluit/open'] = oslist
```

```python
duurdf.columns = ['Duur', 'Tijd','sluit/open']
duurdf.head()
```

```python
sluit = duurdf[duurdf['sluit/open']  == True]
openen = duurdf[duurdf['sluit/open']  == False]
```
