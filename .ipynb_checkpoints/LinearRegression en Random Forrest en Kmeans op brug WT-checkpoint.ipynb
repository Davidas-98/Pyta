## Zet de datums om naar pandas datetime objecten,  zodat er mee te werken is

```python
# voor het =teken staat waar je de data naar wegschrijft, dat is hier de eerste kolom
# na het =teken staat de kolom die je om wil zetten
df['Tijd'] = pd.to_datetime(df['Tijd'])
```

## Is het een foutmelding?

```python
'''
Hier maak je een nieuwe kolom met daarin een int van 0 of 1 of het een foutmelding is

Deze ga je gebruiken om het aantal foutmeldingen per dag te tellen
'''
df['Storing'] = (df.Type == "CAME").astype('int')
```

# Nieuw dataframe maken voor het model

```python
'''
pd.DataFrame maakt een nieuw pandas dataframe aan
groupby groepeert de data per datum
count() telt het aantal meldingen per dag
rename() Tijd kolom naar totaal meldingen zodat de naam duidelijker is


We hebben nu een kleiner dataframe waarin je per datum een rij met daarin verschillende getallen 
waarmee je het model kunt feeden.

De eerste kolom bevat dus het totaal aantal meldingen per dag
''' 
dfdaily = pd.DataFrame(df['Tijd'].groupby(df['Tijd'].dt.date).count()).rename({'Tijd': 'totaal meldingen'}, axis=1)
```

```python
'''
ipv een count doe je hier een sum omdat je alleen de storingen wil tellen die de waarde 1 hebben
Als je hier count doet krijg je het totale aantal meldingen niet alleen de foutmeldingen

De aantal storingen kolom is gevuld met het totaal aantal foutmeldingen per dag
'''
dfdaily['aantal storingen'] = df.groupby(df['Tijd'].dt.date)['Storing'].sum()
```

```python
dfdaily.head()
```

## Plot het aantal storingen tegenover het totaal aantal meldingen

Je zou hier eigenlijk een lineair verband verwachten

```python
plt.figure(figsize=(16,9))
plt.scatter(dfdaily['totaal meldingen'],  dfdaily['aantal storingen'])
```

# Maak het model


## Ordinairy least squares regressie model

```python
LRmodel = LinearRegression()
```

## Random Forrest Ensemble model

```python
# n_estimators is het aantal keer dat hij het model probeert op de data
RFmodel = RandomForestRegressor(n_estimators = 100)
```

## Het model trainen

```python
'''
fit() traint het model met de data die je meegeeft

De eerste param moet een matrix zijn met daarin de kolommen uit je data waarop je het model wil baseren, dus dubbele haken
Dit hoef je niet te snappen doe het gwn :P

Hier train je het model om het aantal storingen te voorspellen aan de hand van het totaal aantal meldingen per dag
'''
LRmodel.fit(dfdaily[['totaal meldingen']],  dfdaily['aantal storingen'])
```

## Het model laten voorspellen

```python
'''
predict() voorspelt dus de kolom die je meegeeft
param is weer een matrix, vraag me niet waarom

We zetten hier de voorspelling in een nieuwe kolom in het dataframe zodat we deze in een grafiek kunnen zetten
en kunnen vergelijken met de echte data
'''
dfdaily['prediction'] = LRmodel.predict(dfdaily[['totaal meldingen']])
```

## Scatterplot van het model

```python
# In deze plot vergelijk je de echte waarden met de waarden die het model voorspelt
plt.figure(figsize=(16,9))
# Scatter waarin de meldingen en het aantal storingen tegen elkaar staan
plt.scatter(dfdaily['totaal meldingen'],  dfdaily['aantal storingen'], color = 'blue')

# Scatter waarin de meldingen en de voorspelling tegen elkaar staan
plt.plot(dfdaily['totaal meldingen'],  dfdaily['prediction'], color = "red")

# opmaak shizzle
plt.title("Linear regression model")
plt.xlabel("Totaal aantal meldingen per dag")
plt.ylabel("aantal foutmeldingen per dag")
plt.legend(('voorspelling','echte waarden'))
```

```python
'''
Hier kijk je naar de functie die uit het model komt om de voorspelling te doen
Deze heeft de vorm y = ax + b

y = het aantal foutmeldingen
x = het totaal antal meldingen

a = de coef_ dit is waar je het totaal aantal meldingen moet vermenigvuldigen met de a om het aantal foutmeldingen te vinden
b  = intercept_ het begin van de lijn 

Dus als je wil weten hoeveel foutmeldingen je hebt bij 400 totale meldingen

dan wordt het 
foutmeldingen = 0.00026338 * 400 + -0.8687718217552445
dit zou -0,76 foutmeldingen voorspellen voor 400 totale meldingen 

'''
LRmodel.intercept_, LRmodel.coef_
```

## Model waarbij je rekening houdt met de dagen van de week

```python
'''
nieuwe kolom met daarin de datums van het type object
'''
dfdaily['datum'] = dfdaily.index
```

```python
'''
een nieuw kolom met daarin de dag van de week 
ranged van 0-6
0 = maandag 
6 = zondag
'''
dfdaily['weekdag'] = pd.to_datetime(dfdaily.datum).dt.dayofweek
```

```python
'''
Per dag een nieuwe kolom met daarin een bool die aangeeft of het die dag van de week is, 
omgezet naar een int om makkelijker mee te  rekenen
'''

dfdaily['maandag'] = (pd.to_datetime(dfdaily.datum).dt.dayofweek == 0).astype('int')
dfdaily['dinsdag'] = (pd.to_datetime(dfdaily.datum).dt.dayofweek == 1).astype('int')
dfdaily['woensdag'] = (pd.to_datetime(dfdaily.datum).dt.dayofweek == 2).astype('int')
dfdaily['donderdag'] = (pd.to_datetime(dfdaily.datum).dt.dayofweek == 3).astype('int')
dfdaily['vrijdag'] = (pd.to_datetime(dfdaily.datum).dt.dayofweek == 4).astype('int')
dfdaily['zaterdag'] = (pd.to_datetime(dfdaily.datum).dt.dayofweek == 5).astype('int')
dfdaily['zondag'] = (pd.to_datetime(dfdaily.datum).dt.dayofweek == 6).astype('int')
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

plt.figure(figsize=(16,9))
kmeans = KMeans(n_clusters=4)
scal = StandardScaler()
colors=['red', 'blue', 'green', 'grey', 'purple']

features = scal.fit_transform(dfdaily[['totaal meldingen','aantal storingen']].astype(np.float64))
y_kmeans = kmeans.fit_predict(features)
plt.scatter(dfdaily['totaal meldingen'], dfdaily['aantal storingen'], c=y_kmeans, s=50, cmap='viridis')

plt.title("K-Means clustering aan de hand van totaal aantal meldingen en aantal foutmeldingen per dag")
plt.xlabel("Totaal aantal meldingen per dag")
plt.ylabel("aantal foutmeldingen per dag")
```
